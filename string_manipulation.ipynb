```python
# 3. For each k-char chunk, move its first character to the end
def moveEachChunkFirstCharToEnd(inputStr: str, k: int) -> str:
    #edge case when both inputs are empty
    if not inputStr or k == 0:
        print('given inputs are empty')
        return inputStr

    #edge case when the k element is greater than input string
    total = len(inputStr)
    result = ''
    if k > total:
        k = k % total

    startIndex = 0
    for i in range(startIndex, total):
        # 0,1,2
        endIndex = startIndex + k

        if endIndex % k == 0 and endIndex < total:

            n = 0
            while k > n:
                if startIndex == 0 or startIndex % k == 0:
                    result = result + inputStr[startIndex]
                else:
                    result = inputStr[startIndex] + result
                startIndex += 1
                n += 1
    return result
```

```python
# 6. Replace each digit in a string with its English name (e.g., '1' → “one”)
def replaceDigitWithString(inputVal: str) -> str:
    if not inputVal:
        return inputVal

    result = []
    englishDigitDic = {
        '0': 'ZERO',
        '1': 'ONE',
        '2': 'TWO',
        '3': 'THREE',
        '4': 'FOUR',
        '5': 'FIVE',
        '6': 'SIX',
        '7': 'SEVEN',
        '8': 'EIGHT',
        '9': 'NINE'
    }
    for val in inputVal:
        if val.isdigit():
            result.append(englishDigitDic.get(val))
        else:
            result.append(val)

    return ' '.join(result)

# using comprehension

print(f"{replaceDigitWithString('12345x')}")
```

```python
# 7. Implement case inversion: lowercase → uppercase and vice versa
def caseInversion(inputVal: str) -> str:
    if not inputVal:
        return inputVal
    
    result = ''
    for val in inputVal:
        if val.islower():
            result += val.upper()
        else:
            result += val.lower()
    return result

# output = caseInversion('Python prograM')
# print(f'{output=}')

def caseInversionWithComprehension(inputVal: str) -> str:

    if not inputVal:
        return inputVal
    
    return ''.join([val.upper() if val.islower() else val.lower() for val in inputVal])

print(f"{caseInversionWithComprehension('pyThon Program')=}")
```

```python
# 8. Compress a string by replacing runs (e.g., “aaabb” → “a3b2”)
def stringCompress(inputVal: str) -> str:

    # edge case
    if not inputVal:
        print('empty input value')
        return inputVal
    
    result = []
    map = {}
    # loop element until end
    for val in inputVal:

        if val in map:
            map[val] = map[val] + 1
        else:
            map[val] = 1
    # print(f'val {map=}')

    for key in map.keys():
        result.append(key)
        result.append(str(map[key]))

    # print(f'{result=}')
    return ''.join(result)


# print(f"{stringCompress('aaabccddddde')=}")
# print(f"{stringCompress('')=}")

def stringCompressWithSingleLoop(inputVal: str) -> str:

    if not inputVal:
        print('empty input')
        return inputVal
    
    result = []
    count = 1

    for key in range(1, len(inputVal)):

        # check with current index with previous index
        if inputVal[key] == inputVal[key - 1]:
            count += 1
        else:
            result.append(inputVal[key-1])
            result.append(str(count))
            count = 1


    result.append(inputVal[-1])
    result.append(str(count))
    return ''.join(result)

output = stringCompressWithSingleLoop('aaabccddddde')
print(f'{output=}')
```

```python
# 9. Expand a compressed string (reverse of #8)
def reverseCompString(inputVal: str) -> str:

    # edge case
    if not inputVal:
        print('empty input value')
        return inputVal
    
    result = []
    loop = 0
    while loop < len(inputVal):

        char = inputVal[loop]

        loop += 1
        numStr = ''

        while loop < len(inputVal) and inputVal[loop].isdigit():
            numStr += inputVal[loop]
            loop += 1

        count = int(numStr) if numStr and numStr != '0' else 1

        result.append(char * count)
    return ''.join(result)

output = reverseCompString('a3b1c2d5e1')
print(f'{output=}')

output1 = reverseCompString('a0')
print(f'{output1=}')

# current code will not work for more than one digits
output2 = reverseCompString('a10b11')
print(f'{output2=}')
```

```python
# 14. Merge two strings by alternating characters

# sample input str1 = "abc" str2 = "xyz"
# Algorithm
# declare variable length = max(len(str1), len(str2)) and result = []
# for loop until meet length check if index < len(str1): result.append str1[index] 
# if index < len(str2): result.append(str2[index])

def mergeTwoStrByAlternateChars(str1: str, str2: str) -> str:

    # if str1 is blank then return str2 directly
    if not str1 and str2:
        return str2
    # if str2 is blank then return str1 directly
    if not str2 and str1:
        return str1

    # if both str1 and str2 are blank then return empty str
    if not str1 and not str2:
        return ''

    # get max len from given two string

    mergeResult = []
    loop = 0
    if len(str1) >= len(str2):
        loop = len(str1)
    else:
        loop = len(str2)

    for index in range(loop):

        if index < len(str1):
            mergeResult.append(str1[index])
        
        if index < len(str2):
            mergeResult.append(str2[index])
    return ''.join(mergeResult)


output1 = mergeTwoStrByAlternateChars('', '')
print(f'{output1=}')

output2 = mergeTwoStrByAlternateChars('abc', '')
print(f'{output2=}')

output3 = mergeTwoStrByAlternateChars('', 'def')
print(f'{output3=}')

output4 = mergeTwoStrByAlternateChars('abcde', 'fgh')
print(f'{output4=}')

output5 = mergeTwoStrByAlternateChars('abc', 'defghijklm')
print(f'{output5=}')

output6 = mergeTwoStrByAlternateChars('a1c', 'd2e')
print(f'{output6=}')

output7 = mergeTwoStrByAlternateChars('123', '456')
print(f'{output7=}')
```

```python
# 16. Check if two strings are anagrams (manual letter counting)

# Algorithm
# after all the edge case validation
#  loop in range(len(str1)): if str1[index] not in str2: return False end of loop return True

def twoAnagramStr(str1: str, str2: str) -> bool:

    # edge case
    # if any input is empty
    if not str1 or not str2:
        return False
    
    # check if both inputs are equal
    if len(str1) != len(str2):
        return False
    
    for index in range(len(str1)):
        if str1[index] not in str2:
            return False
        
    return True


# Test case

result1 = twoAnagramStr('','')
print(f'{result1=}')

result2 = twoAnagramStr('','stet')
print(f'{result2=}')

result3 = twoAnagramStr('stet','')
print(f'{result3=}')

result4 = twoAnagramStr('stet','tests')
print(f'{result4=}')

result5 = twoAnagramStr('testcase','esca')
print(f'{result5=}')

result6 = twoAnagramStr('testcase','escaestt')
print(f'{result6=}')
```

```python
# 17. Implement basic wildcard matching (?=one char, *=any sequence)

# sample input str1 = 'john' str2 = 'j*n'
# sample input str1 = 'john' str2 = 'j?hn'

def wildcardMatch(text: str, pattern: str) -> bool:
    i = 0   # pointer for text
    j = 0   # pointer for pattern
    star_idx = -1      # last position of '*' in pattern
    i_backup = -1      # position in text when '*' was seen

    while i < len(text):
        if j < len(pattern) and (pattern[j] == text[i] or pattern[j] == '?'):
            print(f"MATCH: text[{i}]='{text[i]}' pattern[{j}]='{pattern[j]}'")
            i += 1
            j += 1

        elif j < len(pattern) and pattern[j] == '*':
            star_idx = j
            i_backup = i
            print(f"STAR SEEN: pattern[{j}] is '*', record star_idx={star_idx}, i_backup={i_backup}")
            j += 1  # skip the '*'

        elif star_idx != -1:
            # backtrack: match one more character with '*'
            print(f"BACKTRACK: mismatch at text[{i}]='{text[i]}' and pattern[{j}]='{pattern[j] if j < len(pattern) else 'END'}'")
            j = star_idx + 1
            i_backup += 1
            i = i_backup
            print(f"Retrying: set i={i}, j={j}")

        else:
            print(f"FAIL: no match and no previous '*' to backtrack")
            return False

    # Check for remaining characters in pattern
    while j < len(pattern) and pattern[j] == '*':
        print(f"SKIP STAR: pattern[{j}] is trailing '*', skipping...")
        j += 1

    result = j == len(pattern)
    print(f"END CHECK: pattern pointer at {j}, pattern length {len(pattern)} → {'MATCHED' if result else 'NOT MATCHED'}")
    return result

print("\nCase 1: 'abcde' vs 'a*e'")
print(wildcardMatch("abcde", "a*e"))

print("\nCase 2: 'john' vs 'j*'")
print(wildcardMatch("john", "j*"))

print("\nCase 3: 'john' vs 'j?hn'")
print(wildcardMatch("john", "j?hn"))

print("\nCase 4: 'john' vs 'j*d'")
print(wildcardMatch("john", "j*d"))  # Should fail

print("\nCase 5: 'john' vs '*'")
print(wildcardMatch("john", "*"))

print("\nCase 6: '' vs '*'")
print(wildcardMatch("", "*"))

print("\nCase 7: '' vs '?'")
print(wildcardMatch("", "?"))  # Should fail


# result1 = wildcardMatch('john','j?')
# print(f'{result1=}')

# result2 = wildcardMatch('john','j?oh')
# print(f'{result2=}')

# result3 = wildcardMatch('john','j?hn')
# print(f'{result3=}')

# result4 = wildcardMatch('john','?')
# print(f'{result4=}')

# result5 = wildcardMatch('john','?ohn')
# print(f'{result5=}')

# result6 = wildcardMatch('john','????')
# print(f'{result6=}')

# result7 = wildcardMatch('john','?oh?')
# print(f'{result7=}')
```

```python
# 19. Reverse all words in a sentence (words are space-separated)

# algo
# using two pointer reverse string
# split sentence into list using space then reverse

def reverseWordsInSentence(sentence: str) -> str:

    # edge case
    if not sentence:
        return ''
    
    strList = sentence.split(' ')

    start = 0
    end = len(strList) -1

    while start < end:
        strList[end], strList[start] = strList[start], strList[end]
        start += 1
        end -= 1

    return ' '.join(strList)


result1 = reverseWordsInSentence('')
print(f'{result1=}')

result2 = reverseWordsInSentence('abc')
print(f'{result2=}')

result3 = reverseWordsInSentence('aabc def')
print(f'{result3=}')

result4 = reverseWordsInSentence('Abc def Ghij kkkk')
print(f'{result4=}')

result5 = reverseWordsInSentence('123 456')
print(f'{result5=}')

result6 = reverseWordsInSentence('a b c d')
print(f'{result6=}')

result7 = reverseWordsInSentence('a')
print(f'{result7=}')
```

```python
# 20. Capitalize the first letter of each sentence (detect sentence ends manually)

# Algo
# input = 'welcome to python' output = 'Welcome to python'
# write a common function that accept detect sentence and split them and loop and make first index Capitalize

def capitalizeFirstLetter(input: str) -> str:

    if not input:
        return input
    
    specialChar = ['.', '?', '!']
    maxLen = len(input)
    output = []
    isCapital = False
    for index in range(maxLen):

        if (index == 0 or isCapital == True) and input[index] != ' ':
            if index > 1 and input[index -1] != ' ':
                output.append(' ')
            output.append(input[index].capitalize())
            isCapital = False
        elif input[index] in specialChar:
            output.append(input[index])
            # output.append(' ')
            isCapital = True
        else:
            output.append(input[index])
    return ''.join(output)


result1 = capitalizeFirstLetter('')
print(f'{result1=}')

result2 = capitalizeFirstLetter('hi!johnpaul?')
print(f'{result2=}')

result3 = capitalizeFirstLetter('hi! johnpaul. how are you?')
print(f'{result3=}')

result4 = capitalizeFirstLetter('johnpaul.how are you')
print(f'{result4=}')

result5 = capitalizeFirstLetter('how are you')
print(f'{result5=}')


```

```python
# mock starts with and mock ends with
# input = 'flowerpot' prefix = 'flow'
# loop prefix and check each word is present in input string

def startsWith(input: str, prefix: str) -> bool:

    if not input or not prefix or len(input) < len(prefix):
        return False

    index = 0

    while index < len(prefix):

        if input[index] != prefix[index]:
            return False

        index += 1  

    return True

result = startsWith('flowerpot', 'flower')
print(f'{result=}')

result2 = startsWith('flowerpot', 'floer')
print(f'{result2=}')

result3 = startsWith('flower', 'flowerpot')
print(f'{result3=}')

result4 = startsWith('floer', '')
print(f'{result4=}')

result5 = startsWith('', '')
print(f'{result5=}')

result6 = startsWith('Flowerpot', 'flower')
print(f'{result6=}')
```

```python
def endsWith(input: str, sufix: str) -> bool:

    if not input or not sufix or len(input) < len(sufix):
        return False

    inputIndex = len(input) - 1
    sufixIndex = len(sufix) - 1

    while 0 <= sufixIndex:

        if input[inputIndex] != sufix[sufixIndex]:
            return False

        inputIndex -= 1  
        sufixIndex -= 1

    return True

result = endsWith('flowerpot', 'pot')
print(f'{result=}')

result2 = endsWith('flowerpot', 'kpot')
print(f'{result2=}')

result3 = endsWith('floer', 'flowerpot')
print(f'{result3=}')

result4 = endsWith('floer', '')
print(f'{result4=}')

result5 = endsWith('', '')
print(f'{result5=}')

result6 = endsWith('Flowerpot', 'Pot')
print(f'{result6=}')

result7 = endsWith(' ', ' ')
print(f'{result7=}')
```
