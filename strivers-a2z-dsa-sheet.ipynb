```python
# 02/04/2026
# Step 2 - Sorting - I
# Selection Sort
# Given an array of integers nums, sort the array in non-decreasing order using the selection sort algorithm and return the sorted array.
# A sorted array in non-decreasing order is an array where each element is greater than or equal to all previous elements in the array.
# Input: nums = [7, 4, 1, 5, 3]
# Output: [1, 3, 4, 5, 7]
# Definition
# The Selection Sort algorithm finds the lowest value in an array and moves it to the front of the array. 
# Speed: Selection Sort. The algorithm looks through the array again and again, 
# moving the next lowest values to the front, until the array is sorted.

# Algo
# forloop to track min value
# forloop to goover each element in a array
# Total two forloop
# minIndex to keep mininum value
# two variables i and j to loop over array to find minimum value.
# once inner for loop completed add a condition to check the minimum value from innerloop and swap with minIndex

# nums = [7, 4, 1, 5, 3]
def selectionSort(nums: list[int]) -> list[int]:
    
    for i in range(0, len(nums)):
        currentMinIndex = i
        # print(f'{i} and {nums}')
        for j in range(i, len(nums)):
            if nums[currentMinIndex] > nums[j]:
                currentMinIndex = j
        
        # print(f'before if: {nums[i]} {nums[currentMinIndex]}')
        if nums[i] > nums[currentMinIndex]:
            # print(f'before swap {nums}')
            nums[i], nums[currentMinIndex] = nums[currentMinIndex], nums[i]
            # print(f'after swap {nums}')
    return nums

selection_sort_result = selectionSort([7, 4, 1, 5, 3])
print(f'{selection_sort_result=}')




```

```python
# 02/05/2026
# Step 2 - Sorting - I
# Bubble Sort
# Given an array of integers called nums,sort the array in non-decreasing order 
# using the bubble sort algorithm and return the sorted array.
# Input: nums = [7, 4, 1, 5, 3]
# Output: [1, 3, 4, 5, 7]

# Algo
# 1. n = i  (should be n = 0)
# 2. while n < len(nums)-i-1

# ✓ CORRECTED VERSION:
def bubbleSort(nums: list[int]) -> list[int]:
    
    for i in range(len(nums)):

        n = 0 # always compare with first element
        while n < len(nums) - i - 1:
            if nums[n] > nums[n+1]:
                # print(f'before swap: {nums}')
                nums[n], nums[n+1] = nums[n+1], nums[n]
                # print(f'after swap: {nums}')
            n=n+1
            print(f'after while complete: {nums=}')
        print('while loop')
    return nums

bubbleSort_result = bubbleSort([7, 4, 1, 5, 3])
print(f'{bubbleSort_result=}')
```

```python
# 02/05/2026
# Step 2 - Sorting - I
# Insertion Sort
# Given an array of integers nums, sort the array in non-decreasing order 
# using the recursive Insertion Sort algorithm, and return the sorted array.
# You must implement Insertion Sort using recursion only.

# Input: nums = [7, 4, 1, 5, 3]
# Output: [1, 3, 4, 5, 7]

# Algo
# for loop to go over all the element
# inner while loop to check if previours element is smaller

def insertion_sort(nums: list[int]) -> list[int]:
    print(f'{len(nums)=}')
    for i in range(1,len(nums)):
        print(f'{i=}')
        j = i
        while j > 0 and nums[j-1] > nums[j]:
            print(f'inner loop {i=}')
            nums[j-1], nums[j] = nums[j], nums[j-1]
            j = j-1
    return nums
    
insertion_result = insertion_sort([7, 4, 1, 5, 3])
print(f'{insertion_result=}')


```

```python
# 02/16/2026
# Recursive Bubble Sort

# Algo
# Bubble sort algorithm means loop through an array and compare with each element
# if left > right then we swap left to right and right to left we will do until
# it reaches the end. By doing this the bigger number in an array will go to end
# this is called bubble sort algorithm

class Solution:
    def recursive_bubble_sort(self, nums: list[int], n: int | None = None) -> list[int]:
        """
        Sort `nums` in‑place using bubble sort, but drive the outer loop
        with recursion instead of an explicit `for`/`while`.

        The algorithm works like this:

        1. On each pass we walk through the first `n` elements of `nums`.
           For every pair of neighbours (nums[i], nums[i+1]) we check
           whether the left one is larger than the right one; if it is,
           we swap them.  That way the largest number among the first `n`
           elements moves (“bubbles”) to index `n-1`.

        2. After a pass, the element at position `n-1` is in its final,
           sorted position.  We now only need to sort the prefix of length
           `n-1`, so we call the same function again with `n-1`.

        3. Recursion stops when `n` is 0 or 1 – a list of zero or one
           element is already sorted.

        The parameter `n` is optional so that the first call can simply
        provide the list; the helper logic fills in the length.
        """
        # first call: set n to the length of the list
        if n is None:
            n = len(nums)

        # base case: nothing to sort
        if n <= 1:
            return nums

        # single bubbling pass over nums[0..n-1]
        for i in range(n - 1):           # i = 0, 1, ..., n-2
            # compare adjacent items
            if nums[i] > nums[i + 1]:
                # swap them so the bigger one moves right
                nums[i], nums[i + 1] = nums[i + 1], nums[i]
        # at this point nums[n-1] holds the largest of the first n items

        # recursive step: sort the first n-1 items
        return self.recursive_bubble_sort(nums, n - 1)

# usage example – make sure the class is defined before you instantiate it
s = Solution()
recursive_result = s.recursive_bubble_sort([7, 4, 1, 5, 3], None)
print(f'{recursive_result=}')    # should print [1, 3, 4, 5, 7]
```

```python
# Recursive Insertion Sort

def recursive_insertion_sort(nums: list[int]) -> list[int]:

    # [7, 4, 1, 5, 3]
    def recursive(nums, length):

        # if length is none its not a recursive call hence apply default len(nums)
        if length is None:
            length = 1
        loop = length
        # edge case when length zero or one then its sorted and also break the recursive loop
        if length == len(nums):
            return nums
        
        while loop > 0 and  nums[loop] < nums[loop-1]:
            nums[loop], nums[loop-1] = nums[loop-1], nums[loop] #swap
            loop -=1
        
        print(f'{nums=}')
        recursive(nums, length+1)

    recursive(nums, None)
    return nums

recursive_inserstion_result = recursive_insertion_sort([7, 4, 1, 5, 3])
print(f'{recursive_inserstion_result=}')
```
